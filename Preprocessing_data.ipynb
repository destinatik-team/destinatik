**Data Rating**

```python
rating_df = pd.read_csv('tourism_rating.csv')
rating_df.info()
```

dari hasil diatas terdapat 79 data duplikat perlu dilakukan cleaning

**Sorting Rating**



```python
#place_id 211-334 lalu disort dari yang tertinggi ke rendah pakai rate_df
# Assuming rate_df is your DataFrame and place_id and place_ratings are columns
selected_df = rating_df[(rating_df['Place_Id'] >= 211) & (rating_df['Place_Id'] <= 334)]

# Group by place_id and calculate the mean of ratings
sorted_df = selected_df.groupby('Place_Id')['Place_Ratings'].median().reset_index()

# Sort the DataFrame based on the mean ratings in descending order
sorted_df = sorted_df.sort_values(by='Place_Ratings', ascending=False)
sorted_df
```

```python
print("Jumlah duplikasi: ", rating_df.duplicated().sum())
rating_df.describe()
```
