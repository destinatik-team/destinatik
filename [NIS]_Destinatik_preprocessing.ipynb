## Exploratory Data Analysis (EDA)

**Create new dataframe (exploring destination in Bandung): tourIdBDG_df**

```python
tourIdBDG_df = tourId_df[tourId_df['City']=='Bandung']
tourIdBDG_df
```

**Create new dataframe (rating place only for Bandung): rateBDG_df**

```python
rateBDG_df = rate_df[(rate_df['Place_Id'] >= 211) & (rate_df['Place_Id'] <= 334)]
rateBDG_df
```

**Create new dataframe: rateBDGMed_df**

containing dataframe about rating in Bandung (Median)

```python
rateBDGMed_df = rateBDG_df.groupby(['Place_Id'])['Place_Ratings'].median().reset_index()
rateBDGMed_df
```
